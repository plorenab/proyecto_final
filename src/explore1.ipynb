#### Variables a imputar los Nan

```python
# Zeta Potential (mV) - 68.67% NA, imputación por polímero, fallback, imputar mediana global
col1 = 'Zeta Potential (mV)'
group_col1 = 'Polymer'

# Validar existencia
assert col1 in df1.columns, f"No existe la columna {col1}"
assert group_col1 in df1.columns, f"No existe la columna {group_col1}"

# Forzar numeric y limpiar Polymer
df1[col1] = pd.to_numeric(df1[col1], errors='coerce')
df1[group_col1] = df1[group_col1].astype(str).str.strip()

# Flag de missing original
df1['zeta_missing'] = df1[col1].isna().astype(int)

# Mediana por polymer (NaN para polymers sin datos)
medians_by_poly = df1.groupby(group_col1)[col1].median()

# Mapear la mediana a cada fila (col auxiliar)
df1['zeta_median_by_poly'] = df1[group_col1].map(medians_by_poly)

# Mediana global como fallback
global_median = df1[col1].median()

# Imputación: primero por mediana de polymer, si no -> mediana global
df1[col1] = df1[col1].fillna(df1['zeta_median_by_poly'])
df1[col1] = df1[col1].fillna(global_median)

# Flags de tipo de imputación
df1['zeta_imputed_by_group'] = ((df1['zeta_missing'] == 1) & df1['zeta_median_by_poly'].notna()).astype(int)
df1['zeta_imputed_by_global'] = ((df1['zeta_missing'] == 1) & df1['zeta_median_by_poly'].isna()).astype(int)

# Limpiar columna auxiliar
df1 = df1.drop(columns=['zeta_median_by_poly'])

# Resumen
print("=== Imputación Zeta Potential ===")
print("Total filas:", len(df1))
print("Missing iniciales:", df1['zeta_missing'].sum())
print("Imputadas por mediana de polymer:", df1['zeta_imputed_by_group'].sum())
print("Imputadas por mediana global:", df1['zeta_imputed_by_global'].sum())
print("Missing tras imputación:", df1[col1].isna().sum())
```

```python
# Charge — 66.84% NA, imputación con unknown
df1['charge_missing'] = df1['Charge'].isna().astype(int)
df1['Charge'] = df1['Charge'].fillna('unknown')
```

```python
# Temperature (Avg) — 58.94% NA, in vitro → imputar a 37°C, in vivo → imputar la moda por especie, fallback, imputar mediana global
from IPython.display import display

# Normalizar la columna In Vitro / In Vivo (por si hay variantes)
df1['in_vitro_flag'] = df1['In Vitro or In Vivo?'].astype(str).str.lower().str.contains('in vitro')

# Flag de missing
df1['temp_missing'] = df1['Temperature (Avg)'].isna().astype(int)

# Imputar 37.0 para in vitro (solo donde falta)
mask_invitro_missing = df1['in_vitro_flag'] & df1['Temperature (Avg)'].isna()
df1.loc[mask_invitro_missing, 'Temperature (Avg)'] = 37.0

# Calcular medianas por especie (sin producir warnings)
group_medians = df1.groupby('Species')['Temperature (Avg)'].median()

# Diagnóstico: especies cuya mediana es NaN (especies sin ningún valor)
species_no_temp = group_medians[group_medians.isna()].index.tolist()
if species_no_temp:
    print(f"Especies sin valores de Temperature (Avg): {len(species_no_temp)} (ejemplos): {species_no_temp[:10]}")

# Rellenar con mediana por especie (solo para filas que siguen NaN)
mask_still_missing = df1['Temperature (Avg)'].isna()
df1.loc[mask_still_missing, 'Temperature (Avg)'] = df1.loc[mask_still_missing, 'Species'].map(group_medians)

# Rellenar el resto con mediana global (fallback)
global_median = df1['Temperature (Avg)'].median(skipna=True)
df1['Temperature (Avg)'] = df1['Temperature (Avg)'].fillna(global_median)

print("Imputación completada. Mediana global usada como fallback:", global_median)

# Comprobaciones rápidas
print("Porcentaje NA después de imputación:", df1['Temperature (Avg)'].isna().mean() * 100)
display(df1[['Species','Temperature (Avg)','temp_missing']].head(10))

```

```python
# particles/mL (master) — 47.94% NA, se maneja como “no disponible”
df1['particles_ml_missing'] = df1['particles/mL (master)'].isna().astype(int)
```

```python
# Sex — 45.31% NA, unknown si no está reportado, in vitro: “not_applicable”
df1.loc[df1['In Vitro or In Vivo?'] == 'In Vitro', 'Sex'] = \
    df1.loc[df1['In Vitro or In Vivo?'] == 'In Vitro', 'Sex'].fillna('not_applicable')

df1['Sex'] = df1['Sex'].fillna('unknown')
```

```python
# Surface Chemistry — 19.87% NA /// Polymer Type* — 19.87% NA /// Particle Source* — 19.87% NA /// Chemical Purity — 19.87% NA 
# /// Weathered or Biofouled? — 6.20% NA, todas con unknown
df1['Surface Chemistry'] = df1['Surface Chemistry'].fillna('unknown')
df1['Polymer Type*'] = df1['Polymer Type*'].fillna('unknown')
df1['Particle Source*'] = df1['Particle Source*'].fillna('unknown')
df1['Chemical Purity'] = df1['Chemical Purity'].fillna('unknown')
df1['Weathered or Biofouled?'] = df1['Weathered or Biofouled?'].fillna('unknown')

```

```python
# Particle Volume (μm³) — 6.14% NA, mediana por categoría Size Category

# limpiar y normalizar la columna de categoría (opcional pero recomendable)
df1['Size Category'] = df1['Size Category'].astype(str).str.strip().replace({'nan': np.nan})

# convertir Particle Volume a numérico (coerce convierte texto a NaN)
vol_col = 'Particle Volume (μm^3)'
df1[vol_col] = pd.to_numeric(df1[vol_col], errors='coerce')

# calcular mediana por Size Category
medians_by_size = df1.groupby('Size Category')[vol_col].median()

# mapear la mediana de cada fila según su Size Category
df1['vol_median_by_size'] = df1['Size Category'].map(medians_by_size)

# mediana global como fallback
global_median = df1[vol_col].median()

# crear flag de missing original
df1['particle_volume_missing'] = df1[vol_col].isna().astype(int)

# imputar: primero por mediana de grupo, si no existe usar mediana global
df1[vol_col] = df1[vol_col].fillna(df1['vol_median_by_size'])
df1[vol_col] = df1[vol_col].fillna(global_median)

# limpiar columna auxiliar si quieres
df1 = df1.drop(columns=['vol_median_by_size'])

# Opcional: comprobar que no quedan NaN
print(f"Missing after imputation: {df1[vol_col].isna().sum()}")

```

```python
# Particle Mass (mg) — 6.57% NA, mediana por Polymer
col2 = 'Particle Mass (mg)'
group_col2 = 'Polymer'

# seguridad: columnas existen
assert col2 in df1.columns, f"{col2} no existe"
assert group_col2
# forzar numérico y limpiar polymer strings
df1[col2] = pd.to_numeric(df1[col2], errors='coerce')
df1[group_col2] = df1[group_col2].astype(str).str.strip()

# flag original missing
df1['particle_missing'] = df1[col2].isna().astype(int)

# calcular mediana por polymer (NaN para polymers sin datos)
medians_by_poly = df1.groupby(group_col2)[col2].median()

# mapear la mediana a cada fila
df1['mass_median_by_poly'] = df1[group_col2].map(medians_by_poly)

# mediana global como fallback
global_median = df1[col2].median()

# Imputación: primero por mediana de grupo (si existe), si no -> mediana global
df1[col2] = df1[col2].fillna(df1['mass_median_by_poly'])
df1[col2] = df1[col2].fillna(global_median)

# flags para tipo de imputación
df1['particle_mass_imputed_by_group'] = (
    (df1['particle_missing'] == 1) & df1['mass_median_by_poly'].notna()
).astype(int)

df1['particle_mass_imputed_by_global'] = (
    (df1['particle_missing'] == 1) & df1['mass_median_by_poly'].isna()
).astype(int)

# limpiar columna auxiliar
df1 = df1.drop(columns=['mass_median_by_poly'])

# resumen
print("Total filas:", len(df1))
print("Missing iniciales:", df1['particle_missing'].sum())
print("Imputadas por mediana de polymer:", df1['particle_mass_imputed_by_group'].sum())
print("Imputadas por mediana global:", df1['particle_mass_imputed_by_global'].sum())
print("Missing tras imputación:", df1[col2].isna().sum())

```

```python
# Density (reported/estimated) — 6.44% NA
# nombres de columnas
col_cat = 'Density, reported or estimated'
col_num = 'Density (g/cm^3)'

# asegurarse de que existen
assert col_cat in df1.columns, f"{col_cat} no existe"
assert col_num in df1.columns, f"{col_num} no existe"

# normalizar la columna categórica
df1[col_cat] = df1[col_cat].astype(str).str.strip().str.lower().replace({'nan': np.nan})

# normalizar variantes posibles (por si hay 'estimated ' 'reported ' etc.)
df1[col_cat] = df1[col_cat].replace({
    'est': 'estimated', 'estim': 'estimated',
    'rep': 'reported', 'rpt': 'reported'
})

# inferir 'reported' cuando exista un valor numérico en Density (g/cm^3) pero no haya etiqueta
mask_infer_reported = df1[col_cat].isna() & df1[col_num].notna()
df1.loc[mask_infer_reported, col_cat] = 'reported'

# rellenar el resto de NaN con 'unknown' 
df1[col_cat] = df1[col_cat].fillna('unknown')

# crear flags/binaries útiles
df1['density_reported_flag'] = (df1[col_cat] == 'reported').astype(int)
df1['density_estimated_flag'] = (df1[col_cat] == 'estimated').astype(int)
df1['density_unknown_flag']   = (df1[col_cat] == 'unknown').astype(int)

# convertir a categoría
df1[col_cat] = df1[col_cat].astype('category')

# informe rápido
print(df1[col_cat].value_counts(dropna=False))
print("Flags sums:", df1[['density_reported_flag','density_estimated_flag','density_unknown_flag']].sum().to_dict())
```

```python
# Density (g/cm³) — 0.18% NA, mediana global
df1['Density (g/cm^3)'] = df1['Density (g/cm^3)'].fillna(df1['Density (g/cm^3)'].median())
```

```python
# Particle Length (μm) — 0.31% NA, mediana por shape
# --- Normalizar Shape ---
df1['Shape'] = df1['Shape'].astype(str).str.strip()

# Convertimos "Not Reported" a NA real
df1['Shape'] = df1['Shape'].replace({'Not Reported': np.nan})

# --- Crear flag de missing ---
df1['particle_length_missing'] = df1['Particle Length (μm)'].isna().astype(int)

# --- Calcular medianas ---
median_global = df1['Particle Length (μm)'].median()
median_by_shape = df1.groupby('Shape')['Particle Length (μm)'].median()

# --- Función simple de imputación ---
def impute_particle_length(row):
    if pd.notna(row['Particle Length (μm)']):
        return row['Particle Length (μm)']
    shape = row['Shape']
    if pd.notna(shape) and pd.notna(median_by_shape.get(shape, np.nan)):
        return median_by_shape[shape]
    return median_global

# --- Aplicación ---
df1['Particle Length (μm)'] = df1.apply(impute_particle_length, axis=1)
```

```python
# Sample Size — 12.21% NA, mediana por Experiment Type
import re
# copia de seguridad
if 'Sample Size_original' not in df1.columns:
    df1['Sample Size_original'] = df1['Sample Size']

# función para extraer primer número (enteros o decimales) de una cadena
def extract_first_number(x):
    if pd.isna(x):
        return np.nan
    s = str(x).strip()
    # eliminar comas de millares
    s = s.replace(',', '')
    # buscar primer número (entero o decimal, puede tener decimal with dot)
    m = re.search(r'[-+]?\d*\.?\d+', s)
    if m:
        try:
            return float(m.group(0))
        except:
            return np.nan
    return np.nan

# extraer número
df1['Sample Size_num'] = df1['Sample Size_original'].apply(extract_first_number)

# flag original missing (antes de imputar)
df1['sample_size_missing'] = df1['Sample Size_num'].isna().astype(int)

# mediana por grupo usando transform (alineado con índice)
# primero calculamos mediana por group sólo sobre filas con Experiment Type non-null
# (transform maneja automáticamente grupos vacíos devolviendo NaN para esos índices)
df1['Sample Size_num'] = df1.groupby('Experiment Type')['Sample Size_num'] \
                            .transform(lambda x: x.fillna(x.median()))

# fallback: filas cuya Sample Size_num siga siendo NaN (incluye Experiment Type NaN o grupos sin datos)
global_median = df1['Sample Size_num'].median()
df1['Sample Size_num'] = df1['Sample Size_num'].fillna(global_median)

# opcional: convertir a Int si todos son enteros
if df1['Sample Size_num'].dropna().apply(float.is_integer).all():
    df1['Sample Size_num'] = df1['Sample Size_num'].astype('Int64')

# reemplazar la columna original
df1['Sample Size'] = df1['Sample Size_num']

# limpiar columnas auxiliares si no las necesitas
# df1 = df1.drop(columns=['Sample Size_num'])  # deja Sample Size_original si quieres trazabilidad

print("Global median used:", global_median)
print("Remaining NA in 'Sample Size':", df1['Sample Size'].isna().sum())

# 1. Crear flag para detectar valores imposibles (>100)
df1["sample_size_outlier"] = (df1["Sample Size"] > 100).astype(int)

# 2. Reemplazar valores >100 con NaN temporalmente para imputación correcta
df1.loc[df1["Sample Size"] > 100, "Sample Size"] = np.nan

# 3. Imputación por mediana dentro de cada Experiment Type
df1["Sample Size"] = df1.groupby("Experiment Type")["Sample Size"] \
                        .transform(lambda x: x.fillna(x.median()))

# 4. Fallback global: si aún hay NaN, usar la mediana del dataset
global_median = df1["Sample Size"].median()
df1["Sample Size"] = df1["Sample Size"].fillna(global_median)

# 5. Convertir a entero si todos los valores lo permiten
if df1["Sample Size"].dropna().apply(float.is_integer).all():
    df1["Sample Size"] = df1["Sample Size"].astype("Int64")

print("Mediana global usada en fallback:", global_median)
print("Valores únicos finales:", df1["Sample Size"].unique())
print("Outliers detectados (flag=1):", df1["sample_size_outlier"].sum())
```

```python
# =====================================================
# AÑADIR VARIABLES DE DOSIS ESTANDARIZADAS AL DATASET
# =====================================================
# ----------------------------
# 1. Buscar columna de μg/mL
# ----------------------------
ug_candidates = [
    'μg/mL (master)', 'μg/mL_master', 'μg/ml (master)', 
    'ug/mL (master)', 'ug/ml (master)', 
    'ug_ml_master', 'µg/mL (master)', 'µg/ml (master)'
]

ug_col = None
for c in ug_candidates:
    if c in df1.columns:
        ug_col = c
        break

print("Columna de masa encontrada:", ug_col)

# Crear dose_mg_per_l (1 μg/mL = 1 mg/L)
if ug_col is not None:
    df1['dose_mg_per_l'] = pd.to_numeric(df1[ug_col], errors='coerce')
else:
    df1['dose_mg_per_l'] = np.nan

# Flag de missing
df1['dose_mg_per_l_missing'] = df1['dose_mg_per_l'].isna().astype(int)


# ----------------------------
# 2. Buscar columna de particles/mL
# ----------------------------
particles_candidates = [
    'particles/mL (master)', 'particles/ml (master)',
    'particles/mL_master', 'particles/ml_master',
    'particles_ml_master', 'particles_ml'
]

part_col = None
for c in particles_candidates:
    if c in df1.columns:
        part_col = c
        break

print("Columna de partículas encontrada:", part_col)

# Crear particles_per_ml
if part_col is not None:
    df1['particles_per_ml'] = pd.to_numeric(df1[part_col], errors='coerce')
else:
    df1['particles_per_ml'] = np.nan

# Flag de missing
df1['particles_per_ml_missing'] = df1['particles_per_ml'].isna().astype(int)


# ----------------------------
# 3. Flag general de dosis
# ----------------------------
df1['dose_info_available'] = (
    df1[['dose_mg_per_l', 'particles_per_ml']]
    .notna()
    .any(axis=1)
    .astype(int)
)

# ----------------------------
# Confirmación
# ----------------------------
print("\nColumnas añadidas correctamente:")
for c in ['dose_mg_per_l', 'dose_mg_per_l_missing',
          'particles_per_ml', 'particles_per_ml_missing',
          'dose_info_available']:
    print("✔", c)
```
