```python
dfraw_ini = pd.read_csv("../data/Raw/Raw_experiments_species.csv")
dfraw_ini.shape
```

```python
nan_pct_ini = dfraw_ini.isna().mean().sort_values(ascending=False)
nan_pct_ini.head(20)
```

```python
num_cols_ini = dfraw_ini.select_dtypes(include="number").columns
cat_cols_ini = dfraw_ini.select_dtypes(include="object").columns

len(num_cols_ini), len(cat_cols_ini)
```

```python
mostly_empty_ini = nan_pct_ini[nan_pct_ini > 0.9]
len(mostly_empty_ini)
```

```python
useful_cols_ini = nan_pct_ini[nan_pct_ini < 0.5].index
df_useful_ini = dfraw_ini[useful_cols_ini]

df_useful_ini.shape
```

```python
nunique_ratio_ini = df_useful_ini.nunique() / len(df_useful_ini)
id_like_ini = nunique_ratio_ini[nunique_ratio_ini > 0.9].sort_values(ascending=False)

id_like_ini
```

```python
nunique_ratio = dfraw.nunique() / len(dfraw)
id_like = nunique_ratio[nunique_ratio > 0.9].sort_values(ascending=False)

id_like
```
