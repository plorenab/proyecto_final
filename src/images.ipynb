```python
import requests
headers = {"Authorization": f"Bearer {TOKEN}"}

# Obtener lista de objetos
params = {"project_id": 1345, "page": 1, "per_page": 10}
r = requests.get("https://ecotaxa.obs-vlfr.fr/api/objects", headers=headers, params=params)
data = r.json()

# Descargar imÃ¡genes de cada objeto
for obj in data.get("results", []):
    obj_id = obj["id"]
    img_url = f"https://ecotaxa.obs-vlfr.fr/api/objects/{obj_id}/image"
    r_img = requests.get(img_url, headers=headers)
    with open(f"{obj_id}.png", "wb") as f:
        f.write(r_img.content)

print(r.status_code)
print(r.json())
```
